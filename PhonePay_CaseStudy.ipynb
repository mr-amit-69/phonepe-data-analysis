```python
# que 2: Exploratory Data Analysis (EDA).
# 2.1: Analyse the transacction trends over the year for each state
# a: Calculate the total number of transactions and total transaction amount for each state over the years. 
# a: Display the results in a tabular format.
# Group by State and Year, then sum transactions and amount
state_year_summary = state_txn_users.groupby(['State', 'Year'])[['Transactions', 'Amount (INR)']].sum().reset_index()
state_year_summary
```

```python
# b: Identify the top 5 states with the highest transaction volumes and the top 5 states with the lowest transaction volumes. 
# b: Display the results.
state_total_txn = state_txn_users.groupby('State')['Transactions'].sum().sort_values(ascending=False)

print("Top 5 States by Transaction Volume:\n", state_total_txn.head(5))
print("\nBottom 5 States by Transaction Volume:\n", state_total_txn.tail(5))
```

```python
# 2.2: identify the most common trxn typein each state and quator
#a. For each state and quarter, determine the most frequent transaction type. 
#a. Display the results in a tabular format.
common_txn_type = state_txn_split.groupby(['State', 'Quarter'])['Transaction Type'].agg(lambda x: x.value_counts().idxmax()).reset_index()
common_txn_type
```

```python
#2.3. determine the device brand with the highest number of registered users in each sate.
#a. identify the device brand with  the highest number of user in each state.
#a. Display the result in tabular form.
device = state_device_data.groupby(['State','Brand'])['Registered Users'].sum().reset_index()

top_device_brand = device.loc[device.groupby('State')['Registered Users'].idxmax()]

top_device_brand
```

```python
#2.4 ceate  a list of the top dist per statebased on poplulation.
# a. for each state, identify the dist with the highest population.
#a. diply the result in a tabluar form.
highest_pop = (
    district_demographics.loc
    [district_demographics.groupby('State')['Population'].idxmax()]
               [['State','District','Population']]
)
highest_pop
```

```python
#b. create a volumn chart depicting the dist wiith highest pop for each state.
import matplotlib.pyplot as plt

plt.bar(highest_pop['State'], highest_pop['Population'],color = 'red')
plt.xticks(rotation = 90)
plt.xlabel('State')
plt.ylabel('District')
plt.title('Dist with Highest PoP per State')
plt.show()
```

```python
#2.5:  calculate the average txn value (ATV) for each state.
#a. compute the average txn value for each state in tabular form.
state_atv = state_txn_users.groupby('State')['ATV (INR)'].mean().reset_index()
state_atv
```

```python
#b. identify the top 5 states with highest ATV and 
#b. the top states with lowest ATV.
sort_atv = state_atv.sort_values(by='ATV (INR)',ascending=False)
display(sort_atv.head(5))
display(sort_atv.tail(5))
```

```python
#2.6 analyse app uusage trends
#a. claculate the total no. of app opens over the year and quarters for each state.
#a. display in tabular form.
app_open = state_txn_users.groupby(['State','Year','Quarter'])['App Opens'].sum().reset_index()
app_open.head(12)
```

```python
#b. identify the trend in app usgae by creating a line plot showing,
#   the no. of app open over the time for a selected state."Haryana"
state_app_opens = app_open[app_open['State'] == 'Haryana']

plt.plot(
    state_app_opens['Year'].astype(str) + "-Q" + state_app_opens['Quarter'].astype(str),
    state_app_opens['App Opens'], color = 'orange',
    marker='o'
)

plt.xticks(rotation=45)
plt.title(f"App Opens Trend in Haryana")
plt.xlabel("Year-Quarter")
plt.ylabel("App Opens")

plt.show()
```

```python
#2.7: distribute the txn type
#a. create a bar chart showing the distribution of diff txn type for each
# state for the most recent Quartor in tthe dataset..
latest_year = state_txn_split['Year'].max()
latest_quarter = state_txn_split[state_txn_split['Year'] == latest_year]['Quarter'].max()


latest_data = state_txn_split[(state_txn_split['Year'] == latest_year) &
                              (state_txn_split['Quarter'] == latest_quarter)]

pivot_data = latest_data.pivot_table(index='State',
                                     columns='Transaction Type',
                                     values='Transactions',
                                     aggfunc='sum').fillna(0)

pivot_data.plot(kind='bar', stacked=True, figsize = (12,6))

plt.title(f"Transaction Type Distribution (Year {latest_year}, Quarter {latest_quarter})")
plt.xlabel("State")
plt.ylabel("Transactions")
plt.legend(title="Transaction Type")
plt.xticks(rotation=90)

plt.show()
```

```python
#Que3: Data Quality Chhecks
#3.1: Ensure data consistency across state and district levels
#a. For each state calculate the total nnum of txn,  total txn amt, 
#   and total registered users by summing up the values form the dist level data.
dist_total = district_txn_users.groupby('State')[['Registered Users','Transactions','Amount (INR)']].sum().reset_index()
dist_total
```

```python
#b. compare the result with the correspondign values at the state level to ensure they match.
state_total = state_txn_users.groupby('State')[['Registered Users','Transactions','Amount (INR)']].sum().reset_index()
state_total
```

```python
#c. display any discrepency found b/t the dist levevl and state level
state_dist_comp  = pd.merge(state_total,dist_total,on='State', suffixes = ('_state','_dist'))
state_dist_comp['Users_diff'] = state_dist_comp['Registered Users_state'] - state_dist_comp['Registered Users_dist']
state_dist_comp['Txn_Difference'] = state_dist_comp['Transactions_state'] - state_dist_comp['Transactions_dist'] 
state_dist_comp['Amount_Difference'] = state_dist_comp['Amount (INR)_state'] - state_dist_comp['Amount (INR)_dist']
state_dist_comp
```

```python
#4. Data merging and advance analysis
#4.1: Ratio of users to poplulate by state
#a. Merge the state txn users dataset with dist demographics dataset to calculate,
#   the  ration of registered userd to the populationn for each state. in tabular form
state_total_user = state_txn_users.groupby('State')['Registered Users'].sum().reset_index()
state_total_user
```

```python
#a. 
state_population = district_demographics.groupby("State")["Population"].sum().reset_index()
state_population
```

```python
#a. 
merged_data = pd.merge(state_total_user, state_population, on="State")
merged_data["User_Pop_Ratio"] = (
    merged_data["Registered Users"] / merged_data["Population"]
)

merged_data
```

```python
#b. Create a column chart depicting the ratio of users to population by state
plt.figure(figsize=(12,6))

plt.bar(merged_data["State"], merged_data["User_Pop_Ratio"], color = 'red')

plt.xticks(rotation=90)
plt.title("Ratio of Users to Population by State")
plt.xlabel("State")
plt.ylabel("User to Population Ratio")

plt.show()
```

```python
#4.2: correlate pop density with txn volume
#a. Merge the dist_txn and user dataset with dist_demographics dataset
dist_txn_demo_merge = pd.merge(district_txn_users,district_demographics, on =['State','District'], how = 'inner')
dist_txn_demo_merge.head(5)
```

```python
#b. calculate the correlation b/t pop density and txn volume
corr_volume = dist_txn_demo_merge['Density'].corr(dist_txn_demo_merge['Transactions'])
corr_volume
```

```python
#c. create a scatter plot to visulize the corr b/t pop density and txn volume.
plt.figure(figsize = (8,6))
plt.scatter(dist_txn_demo_merge['Density'],dist_txn_demo_merge['Transactions'],color = 'red')
plt.xlabel('Pop Density')
plt.ylabel('Transactions')
plt.title('Population Density vs Transaction Volume')
plt.show()
```

```python
#4.3: average txn amt per user.
#a. merge relevent dataset to calculate the avg txn amt per user for each sate.

avg_txn_amt = state_txn_users[(state_txn_users["Year"] == latest_year) &
    (state_txn_users["Quarter"] == latest_quarter)
][["State", "Amount (INR)", "Registered Users"]]
avg_txn_amt['Avg_txn_amt_per_user'] = avg_txn_amt['Amount (INR)']/avg_txn_amt['Registered Users']

avg_txn_amt
```

```python
#b. identify the top 5 state with highest avg txn amt per user and the top 5 state,
# with he lowest avg amt txn per user.
top_5_highest = avg_txn_amt.sort_values( by = 'Avg_txn_amt_per_user', ascending = False)

top_5_highest.head(5)
```

```python
top_5_lowest = avg_txn_amt.sort_values( by = 'Avg_txn_amt_per_user', ascending = True)
top_5_lowest.head(5 )
```
